```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from bs4 import BeautifulSoup
import pandas as pd
from selenium.common.exceptions import NoSuchElementException
```

```python
webdriver_path = 'C:\chromedriver-win64\chromedriver.exe'
```

```python
with open('Amazon.html', 'r', encoding='utf-8') as file:
    html_content = file.read()
```

```python
#create a BeautifulSoup object from HTML conten
soup = BeautifulSoup(html_content, 'html.parser')
soup
# Find all divs with class "a-section a-spacing-base"
divs_to_click = soup.find_all('div', class_='a-section a-spacing-base')
divs_to_click
```

```python
options = webdriver.ChromeOptions() # class in Selenium is used to configure options for the Chrome browser
options.add_argument('--headless')  # Run Chrome in headless mode (without GUI)

```

```python
# Create a Chrome browser instance with the specified options
driver = webdriver.Chrome(options=options)
```

```python
# Create empty lists to store data
product_names = []
product_prices = []
product_ratings = []
seller_names = []



for div in divs_to_click:
    # Extract the link from the div
    link = div.find('a')['href']
    print(link)

    # Open the link in the WebDriver
    driver.get(link)

    try:
        # Extract data from the page
        product_name = driver.find_element(By.XPATH, "//span[@class='a-size-large product-title-word-break']").text
        product_price = driver.find_element(By.XPATH, "//span[@class='a-price-whole']").text
        product_rating = driver.find_element(By.XPATH, "//span[@id='acrCustomerReviewText']").text
        seller_name = driver.find_element(By.XPATH, "//div[@id='merchant-info']").text

        # Append data to lists
        product_names.append(product_name)
        product_prices.append(product_price)
        product_ratings.append(product_rating)
        seller_names.append(seller_name)

    except NoSuchElementException as e:
        # Handle the case where an element is not found
        print("Error:", e)

# Close the WebDriver 
driver.quit()

# Create a DataFrame using the lists
data = {
    'product_names': product_names,
    'product_prices': product_prices,
    'product_ratings': product_ratings,
    'seller_names': seller_names
}

df = pd.DataFrame(data)
print(df)
```
